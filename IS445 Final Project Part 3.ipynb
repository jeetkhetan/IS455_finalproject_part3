```python
df = cars_cleaned[cars_cleaned['Car Make'].isin(['Audi','BMW'])]
```

```python
df = df[['Car Make','City MPG','Highway MPG','Horsepower', 'Torque', 'Classification']]
```

```python
df.loc[df['Car Make'] == 'Audi', 'model_no'] = 1
df.loc[df['Car Make'] == "BMW", 'model_no'] = 2
```

```python
df['model_no'] = df['model_no'].astype(int)
```

```python
fig = px.parallel_coordinates(df, color="model_no", labels={"Car Make": "Model",
                "City MPG": "city_mileage", "Highway MPG": "highway_mileage",
                "Horsepower": "Horsepower", "Torque": "Torque", "Classification": "Classification", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
```

```python
fig.show("png")
```

The above parallel co-ordinate helps us nderstand the different characteristics among a model. It can also help us compare models with each other. 

Each model_no is tagged with one car brand.

We can select the likes to select certain values on the co-ordiante to make it interactive. 

There was scope to improve it further if time permitted to add a dropdown filter to select a particular brand and find features for that brand. We could have added a check box dropdown filter to select more than one brand to compare models.
